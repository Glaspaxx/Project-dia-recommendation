```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the MovieLens dataset
url = 'https://files.grouplens.org/datasets/movielens/ml-25m.zip' 
ratings = pd.read_csv(f'data/{url.split("/")[::-1][0].replace(".zip","")}/ratings.csv')
movies = pd.read_csv(f'data/{url.split("/")[::-1][0].replace(".zip","")}/movies.csv')

# Merge ratings and movies data for a comprehensive view
movie_ratings = pd.merge(ratings, movies, on='movieId')

# Set the style for the plots
sns.set(style='whitegrid')
```

```python
# Distribution of Ratings
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=ratings, palette='viridis')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Top 10 Most Rated Movies
plt.figure(figsize=(12, 6))
top_rated_movies = movie_ratings.groupby('title')['rating'].count().sort_values(ascending=False).head(10)
top_rated_movies.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Rated Movies')
plt.xlabel('Movie Title')
plt.ylabel('Number of Ratings')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Average Rating per Movie
plt.figure(figsize=(12, 6))
average_rating_per_movie = movie_ratings.groupby('title')['rating'].mean().sort_values(ascending=False)
sns.histplot(average_rating_per_movie, bins=30, color='salmon', kde=True)
plt.title('Average Rating per Movie')
plt.xlabel('Average Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Number of Ratings per Movie
plt.figure(figsize=(12, 6))
num_ratings_per_movie = movie_ratings.groupby('title')['rating'].count()
sns.histplot(num_ratings_per_movie, bins=50, color='green', kde=True)
plt.title('Number of Ratings per Movie')
plt.xlabel('Number of Ratings')
plt.ylabel('Count')
plt.show()
```
